# Functions

```python
def valueFilter(dataFrame,columnName,values):
    return dataFrame[~dataFrame[columnName].isin(values)]
```

# Formating Data

```python
import pandas as pd
from scipy.stats import pearsonr
import numpy as np
from sklearn.model_selection import train_test_split
```

## drop unecessary features

```python
audi_raw = pd.read_csv('data/audi.csv')
audi_raw = audi_raw.drop(columns=['tax', 'mpg','engineSize'])
audi_raw
```

## filter results to just automatic petrols 

```python
#audi_raw = audi_raw[audi_raw.transmission !='Manual']
#audi_raw = audi_raw[audi_raw.fuelType != 'Diesel']

audi_filtered = valueFilter(audi_raw,"transmission",['Manual','Semi-Auto'])
audi_filtered = valueFilter(audi_filtered,"fuelType",['Diesel','Hybrid'])
audi_filtered= audi_filtered.drop(columns=['transmission','fuelType'])
audi_filtered =audi_filtered[audi_filtered.model !='Q3']
audi_filtered

```
